## X and Y labels

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.title("Distance Travelled Everyday")
plt.xlabel('Speed in km/h', color = 'red')
plt.ylabel('Time in hours', color = 'blue')
plt.show()
```

## Matplot Grid

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker = '.')
plt.title("Distance Travelled Everyday")
plt.xlabel('Speed in km/h', color = 'red')
plt.ylabel('Time in hours', color = 'blue')
plt.grid()
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker = '.')
plt.title("Distance Travelled Everyday")
plt.xlabel('Speed in km/h', color = 'red')
plt.ylabel('Time in hours', color = 'blue')
plt.grid(axis = 'y')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker = '.')
plt.title("Distance Travelled Everyday")
plt.xlabel('Speed in km/h', color = 'red')
plt.ylabel('Time in hours', color = 'blue')
plt.grid(axis = 'y', linestyle = '--', linewidth = 0.50)
plt.show()
```

## Show multiple plots

```python
#plot 1
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(xpoints, ypoints)

#plot 2

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(xpoints, ypoints)

plt.grid(axis = 'x', linestyle = '--', linewidth = 0.50)
plt.show()
```

## Matplotlib Bars

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Student1", "Student2", "Student3", "Student4"])
y = np.array([90, 85, 86, 99])

plt.bar(x,y, color = 'b')
plt.show()
```

Matplotlib Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

g = np.array([35, 25, 25, 15])
grader = ["Grade 1 = 35%", "Grade 2 = 25%", "Grade 3 = 25%", "Grade 4 = 15%"]

plt.legend(title = 'Grader', labels = grader)
plt.pie(g, labels = grader)
plt.show()

```
