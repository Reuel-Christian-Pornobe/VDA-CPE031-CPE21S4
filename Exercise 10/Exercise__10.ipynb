```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])

slope, intercept = np.polyfit(x, y, 1)

y_pred = slope * x + intercept

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([60, 90, 70, 85, 80, 65, 75, 60, 75, 70])
y = np.array([70, 95, 70, 75, 90, 75, 75, 60, 80, 65])

slope, intercept = np.polyfit(x, y, 1)

y_pred = slope * x + intercept

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('Proficiency Exam Score ')
plt.ylabel('Course Grade')
plt.title('Relationship of Exam Score and Course Grade')
plt.show()
```
