```python
#Read the Contents of SampleData.csv

import pandas as pd

df = pd.read_csv('SampleData.csv')

print(df)
```

```python
#Show all top 10 records

print(df.head(10))
```

```python
#Show all bottom 10 records

print(df.tail(10))
```

```python
print(df.duplicated())
df.drop_duplicates(inplace = True)
```

```python
#Eliminate all duplicates
import tabulate
from tabulate import tabulate

df.drop_duplicates(inplace = True)

print(tabulate (df, headers = 'keys', tablefmt = 'psql', showindex=False))
```

```python
#save the contents(removed duplicate records) into a new file, 'SampleDataNew.csv)

df.to_csv('SampleDataNew.csv', index=False)
df = pd.read_csv('SampleDataNew.csv')
print(tabulate (df, headers = 'keys', tablefmt = 'psql', showindex=False))
```

```python
#7 Display the 1st and 2nd columns
df = df.drop(columns = ['Rep', 'Item', 'Units', 'Unit Cost', 'Total' ])
#print(df[['OrderDate', 'Region']])

print(tabulate (df, headers = 'keys', tablefmt = 'psql', showindex=False))
```
