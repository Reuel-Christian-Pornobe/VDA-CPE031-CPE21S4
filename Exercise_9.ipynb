### 2. Read the file using Pandas

```python
import pandas as pd
import csv

df = pd.read_csv('Technological-Products-Sample-Data.csv')


print(df)
```

### 3. Analyze the data and generate the appropriate plot including its properties:

a. Read each device's date of release and sales. Show it using line plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = df.loc[:,'Date of Release']
y = df.loc[:,'Price (USD)']

plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('Year')
plt.ylabel('Sales (USD)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()

plt.show()
```

b. Show line plot with the follwoing style properties:

```python
import matplotlib.pyplot as plt
import numpy as np

x = df.loc[:,'Date of Release']
y = df.loc[:,'Price (USD)']

plt.plot(x,y, marker = 'o', color = 'r', mfc = 'r', linestyle = 'dashed', \
         linewidth = 3.0)
plt.title('Line Plot')
plt.xlabel('Year')
plt.ylabel('Sales (USD)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()

plt.show()
```

C. Read all Product sales data and show it using multiline plot.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Convert 'Date of Release' to date time
df['Date of Release'] = pd.to_datetime(df['Date of Release'])

# Multiple PLots
fig, ax = plt.subplots()

# Iterate through the devices and plot data
for device in  df['Device']:
    device_data = df[df['Device'] == device]

    x_points = np.array(device_data['Date of Release'])
    y_points = np.array(device_data['Price (USD)'])

    ax.plot(x_points, y_points, marker='o', label=device)

#Labels
ax.set_title('Multiline Plot of Product Prices')
ax.set_xlabel('Date of Release')
ax.set_ylabel('Price (USD)')

# Legend
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

d. Read each brand sales data and show it using the bar chart plot.

```python
import matplotlib.pyplot as plt
import numpy as np


x = df.loc[:,'Brand']
y = df.loc[:,'Price (USD)']


plt.bar(x,y, color = 'b')

#Labels
plt.xlabel('Brand')
plt.ylabel('Sales (USD)')

#Layouts
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

e. Read each device and show it using the histogram to see the most common
Country of Origin.

```python
import matplotlib.pyplot as plt
import numpy as np

data = df.loc[:,'Country of Origin']

plt.hist(data, bins=15)
#Labels

plt.title('Histogram of Most Common Country of Origin')
plt.xlabel('Countries')
plt.ylabel('Frequency')
#Layouts

plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```


f. Create a pie chart that shows each brand Country of Origin. Label them.

```python
import matplotlib.pyplot as plt

country_counts = df['Country of Origin'].value_counts() # tallies all the same varaible names together

plt.figure(figsize=(6, 6)) # Size of the circle
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90) # create a pie with a format of 1 decimal percentage

#Labels
plt.title('Country of Origin Distribution')
plt.axis('equal')
plt.show()
```
