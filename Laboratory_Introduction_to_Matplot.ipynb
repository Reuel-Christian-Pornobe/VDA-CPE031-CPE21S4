```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4]
y = [20, 20, 25, 30]

plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [20, 20, 25, 30]

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4]

plt.hist(data)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
data = [[1.5]*10 + [2]*10 + [3]*10]
plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

## Bar Chart

```python
import matplotlib.pyplot as plt

categories = ['Rape', 'Kidnap And Assault', 'Dowry Deaths','Assault against women',"Assault against modesty of women"," Domestic violence","Women Trafficking"]
values =  [df['Rape'].sum(), df['Kidnap And Assault'].sum(), df['Dowry Deaths'].sum(), df['Assault against women'].sum(), df['Assault against modesty of women'].sum(),
           df['Domestic violence'].sum(), df['Women Trafficking'].sum()]

plt.bar(categories, values)
plt.title('Crimes on Women, India(2001)')
plt.xlabel('Category')
plt.ylabel('Value')
# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45, ha='right')
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

## Pie Graph

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('crimesOnWomen.csv' )

crime_columns = ['Rape', 'Kidnap And Assault', 'Dowry Deaths', 'Assault against women',
                 'Assault against modesty of women', 'Domestic violence', 'Women Trafficking']

df['Total Crimes'] = df[crime_columns].sum(axis=1)


plt.figure(figsize=(10, 7))
plt.pie(df['Total Crimes'], labels=df['State'], autopct='%1.1f%%', startangle=140)
plt.title('Total Crimes Against Women by State')
plt.axis('equal')
plt.show()
```

Boxplots

```python
import pandas as pd
import matplotlib.pyplot as plt

#
df = pd.read_csv('crimesOnWomen.csv')

# List of crime categories
crime_columns = ['Rape', 'Kidnap And Assault', 'Dowry Deaths', 'Assault against women',
                 'Assault against modesty of women', 'Domestic violence', 'Women Trafficking']

# Create a boxplot
plt.figure(figsize=(12, 8))
df[crime_columns].boxplot()
plt.title('Distribution of Different Types of Crimes Across States')
plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability
plt.ylabel('Number of Crimes')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('crimesOnWomen.csv')


state_domestic_violence = df.groupby('State')['Domestic violence'].sum()


plt.figure(figsize=(12, 8))
plt.bar(state_domestic_violence.index, state_domestic_violence.values, color='skyblue')
plt.xlabel('State')
plt.ylabel('Domestic Violence Cases')
plt.title('Domestic Violence Cases by States')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
