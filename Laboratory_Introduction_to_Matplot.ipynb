```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4]
y = [20, 20, 25, 30]

plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [20, 20, 25, 30]

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4]

plt.hist(data)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
data = [[1.5]*10 + [2]*10 + [3]*10]
plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# prompt: matplotlib how to plot bar graph using csv file (using the row as the category and the values as columns)

import matplotlib.pyplot as plt
import pandas as pd

# Replace 'your_file.csv' with the actual path to your CSV file
data = pd.read_csv('your_file.csv')

# Assuming your CSV has a header row with column names
categories = data['Your_Row_Column_Name']  # Replace 'Your_Row_Column_Name' with the column name for your categories
values = data.iloc[:, 1:]  # Assuming values start from the second column

# Transpose the data for plotting
values = values.transpose()

# Create the bar graph
plt.figure(figsize=(10, 6))
ax = values.plot(kind='bar', rot=0)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Graph from CSV')
plt.xticks(range(len(categories)), categories)  # Set x-axis labels
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

## Bar Chart

```python
import matplotlib.pyplot as plt

categories = ['Rape', 'Kidnap And Assault', 'Dowry Deaths','Assault against women',"Assault against modesty of women"," Domestic violence","Women Trafficking"]
values =  [df['Rape'].sum(), df['Kidnap And Assault'].sum(), df['Dowry Deaths'].sum(), df['Assault against women'].sum(), df['Assault against modesty of women'].sum(),
           df['Domestic violence'].sum(), df['Women Trafficking'].sum()]

plt.bar(categories, values)
plt.title('Crimes on Women, India(2001)')
plt.xlabel('Category')
plt.ylabel('Value')
# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45, ha='right')
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
