### **Creating a DataFrame**

```python
import pandas as pd
import csv

df = pd.read_csv('heart_2022_no_nans.csv')


print(df)
```

### **1. Which State has the highest amount of people with Heart Disease?**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


state_counts = df['State'].value_counts()

plt.figure(figsize=(12, 8))
plt.bar(state_counts.index, state_counts.values)

plt.title('People with Heart Disease by State')
plt.xlabel('States')
plt.ylabel('Number of People with Heart Disease')

plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

The state with the highest amount of people with Heart Disease is Washington, while the state with the lowest amount of people with Heart disease is Virgin Islands. This information tells us which states need a priority to increase heart care, which is the Washington State.

### **2. Is there a relationship between Body Mass Index and Heart Attack**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# category creation for bmi
bmi_bins = [0, 25, 30, 35, 40, float('inf')]  # Example ranges, adjust as needed
bmi_labels = ['Underweight', 'Healthy', 'Overweight', 'Obese', 'Severely Obese']

# column for bmi category
df['BMI_Category'] = pd.cut(df['BMI'], bins=bmi_bins, labels=bmi_labels)

# grouping bmi category and heart attack occurences
bmi_heart_attack_counts = df.groupby(['BMI_Category', 'HadHeartAttack'])['HadHeartAttack'].count().unstack()


bmi_heart_attack_counts.plot(kind='bar', figsize=(10, 6))

plt.title('Relationship between BMI and Heart Attack Cases')
plt.xlabel('BMI Category')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Heart Attack')

plt.show()
```

The amount of people with Heart disease with the category of Healthy Body Mass Index is surpisingly the majority. Furthemore, it tells us that people with healthy BMI is the majority for heart attack cases. The graph shows that BMI does not have a significant relationship with Heart attack cases.
