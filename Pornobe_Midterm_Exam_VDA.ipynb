## Creating a DataFrame

```python
import pandas as pd
import csv

df = pd.read_csv('heart_2020_cleaned.csv')


print(df)
```

### **Is smoking a strong factor for Heart Disease?**

```python
import matplotlib.pyplot as plt
heart_disease_counts = df['HeartDisease'].value_counts()

smoking_counts = df.groupby(['Smoking', 'HeartDisease'])['HeartDisease'].count().unstack()
smoking_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Heart Disease by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Heart Disease')
plt.show()
```

According to the bar graph, Smokers have a higher percentage to have a heart disease compared to non-smokers. However, it does not show a huge difference between smokers and non-smokers.

### **Which Age group has the highest amount of Heart Disease?**

```python
import pandas as pd
import

heart_disease_counts = df['HeartDisease'].value_counts()

age_category_heart_disease = df.groupby(['AgeCategory', 'HeartDisease'])['HeartDisease'].count().unstack()
age_category_heart_disease.plot(kind='bar', figsize=(12, 6))
plt.title('Heart Disease by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Heart Disease')
plt.show()
```

According to the chart, heart disease starts to show significant increase from age 50-54 to 80 years or older. The highest age category with Heart Disease is 80 or older while the least is 18-24.
