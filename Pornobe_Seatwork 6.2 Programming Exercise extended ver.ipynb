```python
mask = df.columns.str.contains('id$|store_and_fwd_flag', regex=True)
columns_to_drop = df.columns[mask]
columns_to_drop
```

```python
df = df.drop(columns = columns_to_drop)
df.head()
```

```python
df = df.rename(columns = {'tpep_pickup_datetime': 'pickup', 'tpep_dropoff_datetime': 'deopoff'})  # wrong rename
df.columns
```

```python
df = df.rename(columns = {'deopoff':'dropoff'})          #recorrected the wrong name
df.columns
```

```python
df.dtypes
```

```python
df[['pickup', 'dropoff']] = df[['pickup', 'dropoff']].apply(pd.to_datetime)        #change the datatype of pickup and dropoff from object to datetime using the apply method/function
df.dtypes
```

```python
df['elapsed_time'] = df['dropoff'] - df['pickup']
df
```

```python
# uses the assign method/function to create new multiple columns instead of the single way
df = df.assign(
    elapsed_time = lambda x: x.dropoff - x.pickup,
    cost_before_tip = lambda x: x.total_amount - x.tip_amount,
    tip_pct = lambda x: x.tip_amount / x.cost_before_tip,
    fees = lambda x: x.cost_before_tip - x.fare_amount,
    avg_speed = lambda x: x.trip_distance.div(x.elapsed_time.dt.total_seconds()/60/60)
)
df
```
